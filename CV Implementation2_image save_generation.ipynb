```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Image path
path_img=["dataset\\training_set", "dataset\\test_set"]
file_path=[]
lable_name=[]
for folder_path in path_img:
    for root,dirt,file in os.walk(folder_path):
        for files in file:
            img_address=os.path.join(root,files)
            file_path.append(img_address)
            lable_name.append(img_address.split('\\')[2])
```

```python
# showing 25 random images

plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    img=cv2.imread(file_path[i])
    img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    img=cv2.resize(img, (128,128))
    plt.xlabel(lable_name[i])
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.tight_layout()
    plt.imshow(img)
```

```python
# save image in the folder "imagesave"
count=0
for i in range(25):
    imge=cv2.imread(file_path[i])
    cv2.imwrite(os.path.join('imagesave',"Image"+str(i)+".jpg"),imge+1)
```

```python
(imageHeight, ImageWidth) = image.shape[:2]
```

```python
left_top = image[0: information_of_y, 0: information_of_x]
imageShow("left_top", left_top)
#plt.imshow(left_top)
```

```python
cv2.circle(image, (205, 170), 30, (255, 0, 0), 2)
imageShow("right eye",image)
```

```python
movementInformation = np.float32([[1, 0, 50], [0, 1, 105]])
updatedImageAfterMovement = cv2.warpAffine(image, movementInformation, (image.shape[1], image.shape[0]))
imageShow("updated Image After Movement", updatedImageAfterMovement)
```

```python
import imutils
updatedImageAfterMovement = imutils.translate(image, 0, -100)
imageShow("updated Image After Movement", updatedImageAfterMovement)
```
